### Question(1) Create 'price_label' and 'busiest_day' columns

#### First try to create a subset and if,then statement

```python
df_subset = ords_prods_merge[:100000]
```

```python
df = ords_prods_merge
```

```python
def price_label(row):
    
  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

### apply the price_label statement

```python
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
ords_prods_merge = ords_prods_merge.drop(columns = ['price_range'])
```

#### Observation: Python suggest .loc method to be more efficient

```python
# Create conditions with loc for high, mid, low range; python says the .loc will be more a more efficient method

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

### Frequency count of the 3 groups in 'price_range_loc'

```python
df['price_range_loc'].value_counts(dropna = False)
```

#### Observation: Frequency count shows 'Mid-range' the most frequent

### Create a For Loop for 'Busiest day, Least busy, Regularly busy' ; put result on 'result'

```python
result = []
for value in df['orders_day_of_week']:
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least busy')
    else:
        result.append('Regularly busy')
```

### Question(2) Change busiest_day column to be 'busiest days' = 2 busiest days, add 2 slowest days
