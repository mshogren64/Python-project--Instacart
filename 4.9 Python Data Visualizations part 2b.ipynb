### Question(5) Check the difference in expenditure (the “prices” column) depending on the hour of the day.

#### Use a small sample for line graph of prices per hour of the day<br>
-create a seed for random generator<br>
-store True/False value in dev object for test of <= .7<br>
-store 70% of sample in df 'big'<br>

```python
np.random.seed(4)
```

```python
dev = np.random.rand(len(cust_ords_prods))<= .7
```

```python
cust_ords_prods.columns
```

```python
df_small = small[['order_hour_of_day', 'prices']]
```

```python
#1. line_2 order hour of day vs. prices
line_2 = sns.lineplot(data = df_small,x = 'order_hour_of_day',y = 'prices')
```

#### Observation: Orders placed at 10am are the lowest priced orders and at 4-5am are the highest priced orders.
