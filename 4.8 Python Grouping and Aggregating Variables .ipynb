### Question(4) Create a loyatly flag label to a user id based on max order value¶<br>
maximum orders the user has made is over 40, then the customer = “Loyal customer.”<br>
maximum orders the user has made is over 10 but less than or equal to 40, then the customer = “Regular customer.”<br>
maximum orders the user has made is less than or equal to 10, then the customer = “New customer.”

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40,'loyalty_flag'] = 'loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

#### Observation: Instacart has more Regular customers than loyal customers based on given criteria.

### Question(6) Create a spending flag for each user based on the average price across all their orders using the following criteria:<br>
If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”<br>
If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
ords_prods_merge['Average_order']= ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge[['user_id','prices','Average_order']]
```

#### Use .loc function to create 'spending_flag' column based on criteria listed above.

```python
ords_prods_merge.loc[ords_prods_merge['Average_order'] < 10, 'Spending_flag'] = 'Low_spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['Average_order'] >= 10, 'Spending_flag'] = 'High_spender'
```

```python
ords_prods_merge['Spending_flag'].value_counts(dropna = False)
```

#### Observation: More customers are Low spenders (spending less than 10) than High spenders (spending greater than 10)

### Question(7) Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.<br>
median of “days_since_prior_order” is higher than 20, then the customer = “Non-frequent customer.”<br>
median is higher than 10 and lower than or equal to 20, then the customer = “Regular customer.”<br>
median is lower than or equal to 10, then the customer = “Frequent customer.”

```python
ords_prods_merge['median_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.loc[ords_prods_merge['median_order'] > 20,'customer_frequency'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_order'] > 10) & (ords_prods_merge['median_order'] <= 20),'customer_frequency'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_order'] <= 10,'customer_frequency'] = 'Frequent customer'
```

```python
ords_prods_merge['customer_frequency'].value_counts(dropna = False)
```

#### Observation:  67% of Instacart customers are considered frequent, 22% are regular and 11% are non-frequent
